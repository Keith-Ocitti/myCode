```python
values = {"Year":dataSet['Year'].mode(),"Publisher":dataSet['Publisher'].mode()}
yr = dataSet['Year'].mode()
pb = dataSet['Publisher'].mode()
yr
pb
```

```python
ds2 = dataSet.fillna(value = values)
```

```python
ds3 = dataSet.fillna({"Year":dataSet['Year'].mode()[0],"Publisher":dataSet['Publisher'].mode()[0]})
ds3.isnull().sum()
```

```python
ds3.describe()
```

```python
x = ds3[["Rank","Year","NA_Sales","EU_Sales","JP_Sales","Other_Sales"]]
x.describe()
```

```python
y = ds3[['Global_Sales']]
y.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
score = regressor.score(x_test,y_test)
print(score)
```
